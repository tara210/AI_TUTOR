```python
class TracerHub:
    def __init__(self, skills):
        # random initial mastery 0.4–0.8
        self.mastery = {s: random.uniform(0.4,0.8) for s in skills}

    def update(self, skill, correct):
        delta = 0.05 if correct else -0.02
        self.mastery[skill] = max(0, min(1, self.mastery[skill]+delta))
        return self.mastery
```

```python
from typing import List, Literal
from groq import Groq

groq_client = Groq(api_key=os.getenv("GROQ_API_KEY"))
LLM_MODEL = "llama-3.3-70b-versatile"     # fast 70 B model

ChatMessage = dict  # simple alias

def chat_completion(messages: List[ChatMessage],
                    temperature: float = 0.3,
                    model: str = LLM_MODEL) -> str:
    """
    Thin wrapper around Groq chat completions.
    """
    resp = groq_client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return resp.choices[0].message.content.strip()
```

```python
BLOOM_PROMPT = """You are an educational psychologist.
Classify the student's utterance by Bloom’s level:
(Remember / Understand / Apply / Analyze / Evaluate / Create)

Utterance: "{u}"
Level:"""

ICAP_PROMPT = """Label the learning behavior by ICAP framework:
(Interactive, Constructive, Active, Passive)

Utterance: "{u}"
ICAP:"""

MISCON_PROMPT = """Detect specific algebra misconceptions.
Return a JSON list of codes, e.g. ["ALG_SIGN_ERR", "ALG_DIV_DISTRIB"]

Utterance: "{u}"
Misconceptions:"""
```

```python
import json
import ast
import re

def extract_label(text: str, choices: list[str]) -> str:
    """
    Safely extract the best match from model output.
    """
    for choice in choices:
        if re.search(rf"\b{re.escape(choice)}\b", text, re.IGNORECASE):
            return choice
    return "Unknown"

def classify_bloom(u: str) -> str:
    resp = chat_completion([{"role": "user", "content": BLOOM_PROMPT.format(u=u)}])
    print("🔍 BLOOM RAW:", resp)
    try:
        obj = json.loads(resp)
        return obj.get("bloom_level", "Unknown")
    except:
        try:
            obj = ast.literal_eval(resp)
            return obj.get("bloom_level", "Unknown")
        except:
            return extract_label(resp, ["Remember", "Understand", "Apply", "Analyze", "Evaluate", "Create"])

def classify_icap(u: str) -> str:
    resp = chat_completion([{"role": "user", "content": ICAP_PROMPT.format(u=u)}])
    print("🔍 ICAP RAW:", resp)
    try:
        obj = json.loads(resp)
        return obj.get("icap", "Unknown")
    except:
        try:
            obj = ast.literal_eval(resp)
            return obj.get("icap", "Unknown")
        except:
            return extract_label(resp, ["Interactive", "Constructive", "Active", "Passive"])

def detect_miscon(u: str) -> list[str]:
    raw = chat_completion([{"role": "user", "content": MISCON_PROMPT.format(u=u)}])
    print("🔍 MISCON RAW:", raw)
    try:
        parsed = json.loads(raw)
        return parsed if isinstance(parsed, list) else []
    except:
        try:
            parsed = ast.literal_eval(raw)
            return parsed if isinstance(parsed, list) else []
        except:
            return []
```

```python
from transformers import pipeline
import torch, functools

device = 0 if torch.cuda.is_available() else -1

@functools.lru_cache(maxsize=1)
def _sent_pipe():
    return pipeline("text-classification",
                    model="cardiffnlp/twitter-roberta-base-sentiment",
                    device=device)

@functools.lru_cache(maxsize=1)
def _emo_pipe():
    return pipeline("text-classification",
                    model="j-hartmann/emotion-english-distilroberta-base",
                    top_k=None, device=device)

def emotion_sentiment(u: str) -> dict:
    s = _sent_pipe()(u)[0]  # e.g., {'label': 'NEGATIVE', 'score': 0.88}

    # FIX: flatten inner list if needed
    emo_raw = _emo_pipe()(u)
    emo_flat = emo_raw[0] if isinstance(emo_raw, list) and isinstance(emo_raw[0], list) else emo_raw

    e = {d["label"]: d["score"] for d in emo_flat}
    return {"sentiment": s, "emotion": e}
```

```python
class ConceptState:
    def __init__(self, mastery=0.5):
        self.mastery = mastery
        self.t = 0

def update_state(state: ConceptState, correct: bool) -> ConceptState:
    state.t += 1
    delta = 0.2 if correct else -0.1
    state.mastery = min(max(state.mastery + delta, 0.01), 0.99)
    return state

_student_cache: dict[tuple[str,str], ConceptState] = {}
def get_state(student_id: str, concept: str) -> ConceptState:
    return _student_cache.setdefault((student_id, concept), ConceptState())
```

```python
def ras_score(affect: dict, miscon: list[str]) -> int:
    base = 10 * len(miscon)
    if affect["sentiment"]["label"] == "NEGATIVE":
        base += 15 * affect["sentiment"]["score"]
    frustration = affect["emotion"].get("frustration",0)
    base += 20 * frustration
    return int(min(base, 100))
```

```python
def analyse(student_id: str, utterance: str,
            concept: str, correct: bool|None=False):
    gaps = trace_prereq_gaps(student_id, concept)

    bloom  = classify_bloom(utterance)
    icap   = classify_icap(utterance)
    miscon = detect_miscon(utterance)
    affect = emotion_sentiment(utterance)

    state  = get_state(student_id, concept)
    state  = update_state(state, bool(correct))

    ras    = ras_score(affect, miscon)

    decision = {
        "need_visual": "ALG_DIV_DISTRIB" in miscon,
        "tone": "encourage" if affect["sentiment"]["label"]=="NEGATIVE" else "neutral"
    }

    return {
        "tags": dict(bloom=bloom, icap=icap,
                     misconceptions=miscon, affect=affect, ras=ras),
        "state": vars(state),
        "decision": decision,
        "prerequisite_gaps": gaps
    }

```

```python
VISUAL_GIF = "https://cdn.mathvisuals.com/perfect_square.gif"

def generate_reply(utterance: str, analysis: dict) -> dict:
    sys = "You are an expert tutor who diagnoses learning gaps and explains prerequisites."
    prereqs = analysis.get("prerequisite_gaps", [])
    missing = "\n".join([f"- {c} (mastery: {m})" for c, m in prereqs])

    usr = f"""Student said: "{utterance}"

Tags: {json.dumps(analysis['tags'], indent=2)}

If mastery is low or RAS is high, identify prerequisite gaps:
{missing}

Respond with:
1. Clarifying Socratic question
2. Direct explanation if needed
3. Motivation if student seems frustrated
4. Suggest to review key prerequisite(s)
"""
    reply = chat_completion([
        {"role": "system", "content": sys},
        {"role": "user", "content": usr}
    ], temperature=0.4)

    assets = [VISUAL_GIF] if analysis["decision"]["need_visual"] else []
    return dict(reply=reply, ras=analysis["tags"]["ras"], assets=assets, gaps=prereqs)

```

```python
def trace_prereq_gaps(student_id: str, concept: str, threshold=0.6) -> list:
    gaps = []
    for prereq in nx.ancestors(concept_graph, concept):
        state = get_state(student_id, prereq)
        if state.mastery < threshold:
            gaps.append((prereq, round(state.mastery, 2)))
    return sorted(gaps, key=lambda x: x[1])  # sort by lowest mastery
```
